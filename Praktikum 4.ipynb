## untuk deteksi gambar, kaya misal menemukan bentuk yg tepat berdasarkan tekstur dll.

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage.feature import graycomatrix, graycoprops
```

```python
img = skimage.data.astronaut()#disini dia udh nyedianin beberapa gambar termasuk astronot, cek di skicit image
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)#HSV 3 dimensi, jadi harus dimanipulasi agar 2 dimensi untuk glcm

fig,axs = plt.subplots(1,2, figsize = (10,10))

ax = axs.ravel()

ax[0].imshow(img)
ax[0].set_title("RGB")

ax[1].imshow(img_hsv, cmap = "gray")
ax[1].set_title("Gray")
```

```python
# Ambil channel V (Value) dari gambar HSV
img_value = img_hsv[:,:,2] #manipulasi dimensi
```

```python
# Ambil channel H (Hue) dari gambar HSV
img_hue = img_hsv[2,:,:]
```

```python
# Ambil channel S (Saturation) dari gambar HSV
img_saturation = img_hsv[:,2,:]
```

```python
mean = np.mean(img_value.ravel()) #mencari rata rata pada matrik gambar, mencari ciri ciri pada suatu gambar
std = np.std(img_value.ravel()) #mencari standar deviasi
#gunanya misal g dapat nilai ambang batas, pake mean std ini, ulik dokumentasi numpy, numpy untuk pemrosesan array dan matriks.
print(mean, std)
```

```python
mean = np.mean(img_hue.ravel()) #mencari rata rata pada matrik gambar, mencari ciri ciri pada suatu gambar
std = np.std(img_hue.ravel()) #mencari standar deviasi
#gunanya misal g dapat nilai ambang batas, pake mean std ini, ulik dokumentasi numpy, numpy untuk pemrosesan array dan matriks.
print(mean, std)
```

```python
mean = np.mean(img_saturation.ravel()) #mencari rata rata pada matrik gambar, mencari ciri ciri pada suatu gambar
std = np.std(img_saturation.ravel()) #mencari standar deviasi
#gunanya misal g dapat nilai ambang batas, pake mean std ini, ulik dokumentasi numpy, numpy untuk pemrosesan array dan matriks.
print(mean, std)
```

```python
glcmvalue = graycomatrix(img_value, distances = [1], angles = [0], levels = 256, symmetric = True, normed = True)
print(glcmvalue)
```

#### grey level co-occurence matrix, salah satu metode untu dapetin yg kode dibawah/ ketidaksamaan korelasi, homogenitas, kontras.
#### mengamati perulangan pola, distribusi parsial dan susunan warna
#### misal pake numpy doang, g bs lihat perulangan pola dll
#### cara kerja cek piksel ketetanggan berdasarkan sudut
#### maksud dari parameter distance, angles, dan level itu cek lagi web hehehehe (itu paling penting cuyyy)
#### symetric = true berarti ururtan nilai pada pasangan piksel tidak dipertimbangakan
#### normed = untuk normalisasi hasilnya agar lebih mudah di ekstrasi fiturnya, kalo true berarti tidak di normalisasi

```python
glcmhue = graycomatrix(img_hue, distances = [1], angles = [0], levels = 256, symmetric = True, normed = True)
print(glcmhue)
```

```python
glcmsaturation= graycomatrix(img_saturation, distances = [1], angles = [0], levels = 256, symmetric = True, normed = True)
print(glcmsaturation)
```

```python
contrastvalue = graycoprops(glcmvalue, 'contrast')[0,0]
dissimilarityvalue = graycoprops(glcmvalue, 'dissimilarity')[0,0]
homogeneityvalue = graycoprops(glcmvalue, 'homogeneity')[0,0]
energyvalue = graycoprops(glcmvalue, 'energy')[0,0]
correlationvalue = graycoprops(glcmvalue, 'correlation')[0,0]
```

```python
contrasthue = graycoprops(glcmhue, 'contrast')[0,0]
dissimilarityhue = graycoprops(glcmhue, 'dissimilarity')[0,0]
homogeneityhue = graycoprops(glcmhue, 'homogeneity')[0,0]
energyhue = graycoprops(glcmhue, 'energy')[0,0]
correlationhue = graycoprops(glcmhue, 'correlation')[0,0]
```

```python
contrastsaturation = graycoprops(glcmsaturation, 'contrast')[0,0]
dissimilaritysaturation = graycoprops(glcmsaturation, 'dissimilarity')[0,0]
homogeneitysaturation = graycoprops(glcmsaturation, 'homogeneity')[0,0]
energysaturation = graycoprops(glcmsaturation, 'energy')[0,0]
correlationsaturation = graycoprops(glcmsaturation, 'correlation')[0,0]
```

#### proses yg diatas tu ada rumusnya guys

```python
print (f'contrast : {contrasthue}')
print (f'dissimilarity : {dissimilarityhue}')
print (f'homogeneity : {homogeneityhue}')
print (f'energy : {energyhue}')
print (f'correlation : {correlationhue}')
```

```python
print (f'contrast : {contrastsaturation}')
print (f'dissimilarity : {dissimilaritysaturation}')
print (f'homogeneity : {homogeneitysaturation}')
print (f'energy : {energysaturation}')
print (f'correlation : {correlationsaturation}')
```
